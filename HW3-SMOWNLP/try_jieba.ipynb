```python
import jieba
import pandas as pd
import gensim
import gensim.models as models
import jieba.analyse
```

```python
file = open('songText.txt',encoding='utf8')
file.readline()
```

```python
lines =list()
for line in file.readlines():
    if line=="\n":
        print("skip")
    elif line.startswith("詞"):
        print("詞")
    elif line.startswith("曲"):
        print("曲")
    elif line.startswith("~"):
        print("~")
    elif line.startswith("-"):
        print("-")
    elif "ps" in line:
        print("ps")
    elif "原曲" in line:
        print("原曲")
    else:
        cutResult=jieba.cut(line)
        goodCutResult=list()
        for n in cutResult:
            if n not in [',','。','!',' ','\u3000','(',')','\n','…','，','。','的','他','！']:
                goodCutResult.append(n)
        lines.append(goodCutResult)
                
                
```

```python
lines[0:100]
```

```python
myModel=models.Word2Vec()
myModel.build_vocab(lines)
myModel.train(lines,total_examples=myModel.corpus_count,epochs=1)
```
