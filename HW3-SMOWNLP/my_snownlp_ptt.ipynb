```python
from PttWebCrawler.crawler import *

c = PttWebCrawler(as_lib=True)
from snownlp import normal
from snownlp import seg
from snownlp.summary import textrank
import json
```

```python
t=c.parse_articles(1, 2, 'gossiping')
```

```python
with open(t, 'r',encoding = 'utf8') as f:
    data=json.load(f)
titles=[]
for article in data['articles']:
    titles.append(article['article_title'])
titles
```

```python
articals=[]
for text in data['articles']:
    articals.append(text['content'])
len(articals)
```

```python
text_in_sentences=[]
for text in articals:
    text=normal.zh2hans(text)
#     
for text in articals:
    text_in_sentences.append(normal.get_sentences(text))
text_in_sentences
```

```python
alldoc = []
for sents in text_in_sentences:
    doc=[]
    for sent in sents:
        words = seg.seg(sent)
        words = normal.filter_stop(words)
        doc.append(words)
    alldoc.append(doc)
```

```python
text_in_sentences[5]
```

```python
i=0
for doc in alldoc:
    print("artical",i,titles[i])
    rank = textrank.TextRank(doc)
    rank.solve()
    for index in rank.top_index(5):
        print(text_in_sentences[i][index])
    i=i+1
```

```python
i=0
for doc in alldoc:
    print("artical",i,titles[i])
    keyword_rank = textrank.KeywordTextRank(doc)
    keyword_rank.solve()
    for w in keyword_rank.top_index(5):
        print(w)
    i=i+1
```
